```python
# 從 Yahoo Finance 中讀取 S&P 500 的歷史價格

df_NQ = yf.download('^IXIC', start='2018-01-01', end='2023-06-28')
df_NQ
```

```python
df_NQ.describe(include='all')
```

```python
df_NQ.info()
```

```python
df_NQ['percent_chnage'] = (df_NQ['Close']/df_NQ['Open']-1)*100
df_NQ
```

```python
df_NQ['Jump'] = np.where((df_NQ.percent_chnage>=0) & (1>df_NQ.percent_chnage) ,1,0)
df_NQ
```

```python
df_NQ['Big_Jump'] = np.where((df_NQ.percent_chnage>=1),1,0)
df_NQ
```

```python
df_NQ['Drop'] = np.where((df_NQ.percent_chnage<0) & (-1<df_NQ.percent_chnage) ,1,0)
df_NQ
```

```python
df_NQ['Big_Drop'] = np.where((df_NQ.percent_chnage<=-1),1,0)
df_NQ
```
